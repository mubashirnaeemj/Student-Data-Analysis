## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
# Replace income values
df['income'] = df['income'].apply(lambda x: 0 if x == '<=50K' else 1)

# Display the updated DataFrame
print(df)
```

## ***18. Which Workclass getting The Highest Salary***

```python
# Replace income values with numerical values for easier calculation
df['income'] = df['income'].apply(lambda x: 0 if x == '<=50K' else 1)

# Group by workclass and calculate the mean income
workclass_salary = df.groupby('workclass')['income'].mean()

# Find the workclass with the highest mean income
highest_salary_workclass = workclass_salary.idxmax()
highest_salary_value = workclass_salary.max()

print(f"The workclass with the highest average salary is: {highest_salary_workclass} with an average income of: {highest_salary_value}")
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
# Check the unique values in the Income column
print("Unique values in Income column:", df['income'].unique())

# Filter for individuals earning >50K
high_income = df[df['income'] == '>50K']

# Check if the filtered DataFrame is empty
if high_income.empty:
    print("No entries found with Income >50K.")
else:
    # Calculate the proportion of high earners by gender
    gender_income = high_income['gender'].value_counts(normalize=True)
    
    # Display the results
    print(gender_income)
    
    # Determine which gender has a higher chance of earning >50K
    better_chance_gender = gender_income.idxmax()
    better_chance_value = gender_income.max()
    
    print(f"{better_chance_gender} have a better chance of earning more than $50K with a proportion of {better_chance_value:.2f}.")
```
